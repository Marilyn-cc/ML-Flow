## Welcome Back

Quick notebook, powerful idea. A set is a collection that automatically keeps only unique values and has no fixed order. If you have ever needed to answer "what are the distinct values in this list?", a set is the tool built exactly for that.


## Why Does This Matter?

Imagine a list of every product purchased today at a shop:

```python
purchases = ["Milk", "Bread", "Milk", "Rice", "Bread", "Milk"]
```

If you want to know which **distinct** products were sold, counting manually is tedious and error-prone. A set gives you the answer instantly, which is a task you will repeat constantly once working with real transaction data.


## Step 1: Creating a Set

A set is created using curly braces `{}`, similar to a dictionary, but without key-value pairs, just individual values.


````
fruits = {"Mango", "Banana", "Pineapple", "Mango", "Banana"}
print(fruits)
print(type(fruits))
````

Notice the duplicates disappeared automatically, and the printed order may not match the order you typed them in. Sets do not preserve order, and that is expected behavior, not a bug.


````
# Converting a list with duplicates into a set removes them instantly
purchases = ["Milk", "Bread", "Milk", "Rice", "Bread", "Milk"]
unique_products = set(purchases)

print(unique_products)
print("Total purchases:", len(purchases))
print("Distinct products:", len(unique_products))
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Creating an empty set with `{}` | `empty = {}` | This actually creates an empty **dictionary** | Use `set()` for an empty set |
| Expecting a set to preserve order | Printing a set and expecting the original order | Order is not guaranteed | Convert to a `list` if order matters |
| Trying to index a set | `my_set[0]` | `TypeError`, sets are not ordered | Convert to a list first: `list(my_set)[0]` |
| Adding a duplicate and expecting it to appear twice | `my_set.add("Milk")` when `"Milk"` already exists | Nothing changes, duplicates are ignored | This is expected set behavior |


````
empty_set = set()   # correct way to create an empty set
print(type(empty_set))

empty_dict = {}      # this is actually a dictionary, not a set!
print(type(empty_dict))
````

## Step 2: Adding and Removing Items

| Method | What it does | Example |
|---|---|---|
| `.add(x)` | Adds a single item | `products.add("Sugar")` |
| `.update(list)` | Adds multiple items | `products.update(["Eggs", "Butter"])` |
| `.remove(x)` | Removes an item, errors if missing | `products.remove("Sugar")` |
| `.discard(x)` | Removes an item, no error if missing | `products.discard("Sugar")` |


````
products = {"Milk", "Bread"}

products.add("Rice")
print(products)

products.update(["Sugar", "Eggs"])
print(products)

products.discard("NotInSet")   # no error, even though it doesn't exist
print(products)
````

## Step 3: Set Operations

This is where sets truly shine: comparing two collections.

| Operation | Symbol | Method | Meaning |
|---|---|---|---|
| Union | `\|` | `.union()` | All items from both sets, combined |
| Intersection | `&` | `.intersection()` | Only items in both sets |
| Difference | `-` | `.difference()` | Items in the first set but not the second |
| Symmetric difference | `^` | `.symmetric_difference()` | Items in either set, but not both |


````
customer_a_purchases = {"Milk", "Bread", "Rice"}
customer_b_purchases = {"Rice", "Sugar", "Eggs"}

print("Union:", customer_a_purchases | customer_b_purchases)
print("Intersection:", customer_a_purchases & customer_b_purchases)
print("Difference (A - B):", customer_a_purchases - customer_b_purchases)
print("Symmetric difference:", customer_a_purchases ^ customer_b_purchases)
````

## Fun Facts

- Sets in Python are built on the same underlying structure as dictionaries, which is why creating an empty one requires `set()` rather than `{}`.
- Checking whether a value exists in a set (`x in my_set`) is typically much faster than checking a list, especially for large collections, because of how sets are stored internally.
- The mathematical concepts of union, intersection, and difference that you just used come directly from set theory, the same branch of mathematics used in relational databases and SQL joins.


## Quick Recap

- Sets store only unique, unordered values, created with `{}` or `set()`.
- Converting a list to a set (`set(my_list)`) is the fastest way to remove duplicates.
- Add items with `.add()` or `.update()`; remove with `.remove()` or `.discard()`.
- Use `|`, `&`, `-`, and `^` to compare two sets: union, intersection, difference, and symmetric difference.
- Sets cannot be indexed, since they have no guaranteed order.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Create a set | `{item1, item2}` | `{"Milk", "Bread"}` |
| Empty set | `set()` | `empty = set()` |
| Remove duplicates from a list | `set(my_list)` | `set([1,1,2])` → `{1, 2}` |
| Add an item | `.add(x)` | `products.add("Rice")` |
| Add multiple items | `.update(list)` | `products.update(["A","B"])` |
| Remove an item | `.remove(x)` / `.discard(x)` | `products.discard("Rice")` |
| Union | `\|` or `.union()` | `a \| b` |
| Intersection | `&` or `.intersection()` | `a & b` |
| Difference | `-` or `.difference()` | `a - b` |
| Symmetric difference | `^` | `a ^ b` |
| Check membership | `x in set` | `"Milk" in products` |
