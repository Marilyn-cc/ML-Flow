## Step 1: Arithmetic Operators

These operators do math, just like you'd expect.

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division (always returns a float) | `5 / 2` | `2.5` |
| `//` | Floor division (rounds down) | `5 // 2` | `2` |
| `%` | Modulus (the remainder) | `5 % 2` | `1` |
| `**` | Exponent (power) | `5 ** 2` | `25` |


````
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
````

### A note on `%` (modulus)

The modulus operator is one beginners often overlook, but it's extremely useful. It gives you the **remainder** after division, which is perfect for checking things like whether a number is even or odd.

```python
10 % 2   # 0 -> even
7 % 2    # 1 -> odd
```


````
number = 7
remainder = number % 2

if remainder == 0:
    print("Even")
else:
    print("Odd")

# Don't worry about the if statement syntax yet — you'll master it in Notebook 9!
````

## Step 2: Comparison Operators

Comparison operators compare two values and always return a **boolean** (`True` or `False`).

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater than or equal to | `5 >= 5` | `True` |
| `<=` | Less than or equal to | `5 <= 4` | `False` |

This is where the earlier warning about `=` vs `==` becomes important: `=` **assigns** a value, `==` **compares** two values.


````
price = 500
budget = 700

print(price == budget)
print(price != budget)
print(price < budget)
print(price >= budget)
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Using `=` instead of `==` | `if price = 500:` | `SyntaxError` | `if price == 500:` |
| Comparing different types | `"5" == 5` | Returns `False` (not an error, but often not what you meant) | Convert types first: `int("5") == 5` |
| Forgetting `/` always gives a float | `10 / 2` expecting `int` | Returns `5.0`, not `5` | Use `//` if you need a whole number |
| Confusing `and` with `&`, `or` with `|` | Works differently on plain values vs. NumPy/Pandas data | Use `and`/`or` for plain Python; you'll meet `&`/`|` in the NumPy notebooks | — |

Every one of these is a completely normal thing to trip over while learning. Reading the error message is your best debugging tool.


## Step 3: Logical Operators

Logical operators combine multiple `True`/`False` conditions.

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `and` | True only if **both** conditions are True | `True and False` | `False` |
| `or` | True if **at least one** condition is True | `True or False` | `True` |
| `not` | Flips True to False, or False to True | `not True` | `False` |


````
in_stock = True
on_sale = False

print(in_stock and on_sale)   # both must be True
print(in_stock or on_sale)    # only one needs to be True
print(not in_stock)           # flips the value
````

## Step 4: Assignment Operators (Shortcuts)

You met one of these briefly in Notebook 1. Here's the full set:

| Operator | Meaning | Equivalent to |
|---|---|---|
| `+=` | Add and reassign | `x = x + value` |
| `-=` | Subtract and reassign | `x = x - value` |
| `*=` | Multiply and reassign | `x = x * value` |
| `/=` | Divide and reassign | `x = x / value` |
| `//=` | Floor divide and reassign | `x = x // value` |
| `**=` | Exponent and reassign | `x = x ** value` |


````
stock = 100

stock -= 15   # 15 items sold
print(stock)

stock += 40   # 40 items restocked
print(stock)

stock **= 1   # stays the same, just demonstrating the operator
print(stock)
````

## Step 5: Operator Precedence (Order of Operations)

Just like in math class, Python follows an order of operations (often remembered as PEMDAS): parentheses first, then exponents, then multiplication/division, then addition/subtraction.


````
result_1 = 2 + 3 * 4       # multiplication happens first
result_2 = (2 + 3) * 4     # parentheses force addition first

print(result_1)
print(result_2)
````

When in doubt, use parentheses. They cost nothing and make your intentions crystal clear, both to Python and to anyone reading your code later, including future you.

## Quick Recap

- Arithmetic operators (`+ - * / // % **`) perform calculations.
- Comparison operators (`== != > < >= <=`) always return a boolean.
- Logical operators (`and`, `or`, `not`) combine or flip boolean conditions.
- Assignment shortcut operators (`+= -= *= /=` etc.) update a variable using its current value.
- Python follows a strict order of operations; use parentheses to make your intent clear.

## Cheat Sheet

| Category | Operators | Notes |
|---|---|---|
| Arithmetic | `+  -  *  /  //  %  **` | `/` always returns `float`; `//` returns whole number |
| Comparison | `==  !=  >  <  >=  <=` | Always returns `True`/`False` |
| Logical | `and  or  not` | Combine or flip boolean conditions |
| Assignment shortcuts | `+=  -=  *=  /=  //=  **=` | Update a variable using its current value |
| Precedence | `()` then `**` then `* /` then `+ -` | Use parentheses to be explicit |
