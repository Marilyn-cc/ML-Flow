## Step 1: The For Loop

A `for` loop runs a block of code once for every item in a collection.

```python
for item in collection:
    # this code runs once per item
```


````
fruits = ["Mango", "Banana", "Pineapple"]

for fruit in fruits:
    print(fruit)
````

On each pass through the loop, the variable `fruit` takes on the value of the next item in the list. You can name this variable anything you like.


````
prices = [150, 65, 300]

for price in prices:
    discounted = price * 0.9
    print(f"Original: {price}, Discounted: {discounted}")
````

## Step 2: Looping With `range()`

Sometimes you want to repeat something a fixed number of times, without looping over an existing collection. `range()` generates a sequence of numbers for exactly this purpose.

| Call | Produces |
|---|---|
| `range(5)` | `0, 1, 2, 3, 4` |
| `range(2, 6)` | `2, 3, 4, 5` |
| `range(0, 10, 2)` | `0, 2, 4, 6, 8` (step of 2) |


````
for i in range(5):
    print(i)
````

````
for i in range(1, 6):
    print(f"Round {i}")
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Forgetting `range()` stops before its endpoint | `range(1, 5)` expecting 5 included | Only produces `1, 2, 3, 4` | Remember the stop value is excluded |
| Forgetting the colon | `for x in items` | `SyntaxError` | `for x in items:` |
| Modifying a list while looping over it | Removing items inside a `for` loop over that same list | Skips items unpredictably | Loop over a copy: `for x in items[:]:` |
| Creating an infinite `while` loop | Forgetting to update the loop's condition variable | Code runs forever | Always update the condition inside the loop |


## Step 3: Looping With Index Access

Sometimes you need both the position and the value. `enumerate()` gives you both at once.


````
fruits = ["Mango", "Banana", "Pineapple"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
````

## Step 4: The While Loop

A `while` loop repeats as long as a condition stays `True`. It is most useful when you do not know in advance how many times you need to repeat something.


````
count = 0

while count < 5:
    print(count)
    count += 1   # this line is essential, or the loop never ends
````

### A serious warning about infinite loops

If you forget to update the condition variable inside a `while` loop, it will run forever, and you will need to manually interrupt the cell (the stop/interrupt button in Jupyter). This is one of the most common beginner mistakes, and everyone runs into it eventually.


````
stock = 20
sold_today = 0

while stock > 0 and sold_today < 5:
    stock -= 1
    sold_today += 1

print("Remaining stock:", stock)
print("Sold today:", sold_today)
````

## Step 5: Controlling Loops with `break` and `continue`

| Keyword | What it does |
|---|---|
| `break` | Immediately exits the loop entirely |
| `continue` | Skips the rest of the current pass, moves to the next item |


````
numbers = [4, 8, 15, 16, 23, 42]

for number in numbers:
    if number == 16:
        break   # stop the loop entirely once we hit 16
    print(number)
````

````
numbers = [4, 8, 15, 16, 23, 42]

for number in numbers:
    if number % 2 != 0:
        continue   # skip odd numbers, move to the next one
    print(number)
````

## Step 6: Nested Loops

A loop can contain another loop inside it. This is useful for grid-like data, such as rows and columns.


````
sizes = ["Small", "Medium", "Large"]
colors = ["Red", "Blue"]

for size in sizes:
    for color in colors:
        print(f"{size} - {color}")
````

## Loops Combined with Conditionals

Loops become truly powerful once combined with `if` statements, letting you process data conditionally, item by item.


````
prices = [150, 65, 300, 90, 220]

for price in prices:
    if price > 100:
        print(f"{price}: expensive")
    else:
        print(f"{price}: affordable")
````

## Fun Facts

- Python's `for` loop is technically iterating over any "iterable," which includes lists, strings, dictionaries, sets, and even file objects, not just numbers.
- `range()` does not actually build a full list of numbers in memory; it generates each number only when needed, which is far more memory-efficient for very large ranges.
- Once you reach the NumPy and Pandas notebooks later in this course, you will often be able to replace explicit loops with faster, vectorized operations, but understanding loops first is essential to understanding why that speed-up matters.


## Quick Recap

- `for` loops iterate over a collection, running once per item.
- `range(start, stop, step)` generates a sequence of numbers to loop over.
- `enumerate()` gives you both the index and the value while looping.
- `while` loops repeat as long as a condition remains `True`; always update the condition to avoid infinite loops.
- `break` exits a loop entirely; `continue` skips to the next iteration.
- Loops can be nested, and combined with `if` statements for conditional processing.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Basic for loop | `for x in collection:` | `for fruit in fruits:` |
| Loop a fixed number of times | `for i in range(n):` | `for i in range(5):` |
| Loop with start/stop/step | `range(start, stop, step)` | `range(0, 10, 2)` |
| Loop with index | `enumerate(collection)` | `for i, x in enumerate(fruits):` |
| While loop | `while condition:` | `while count < 5:` |
| Exit a loop early | `break` | `if x == target: break` |
| Skip to next iteration | `continue` | `if x % 2 != 0: continue` |
| Nested loop | loop inside a loop | `for a in x: for b in y:` |
