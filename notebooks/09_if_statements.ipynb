## Why Does This Matter?

Real data decisions are almost always conditional:

- If a customer's order total is above 2000, waive the delivery fee.
- If a product's stock count is 0, mark it as out of stock.
- If a student's GPA is 3.5 or higher, flag them for honors.

Every one of these is an `if` statement waiting to be written. This is the logic that powers dashboards, alerts, and automated business rules.


## Step 1: The Basic If Statement

An `if` statement runs a block of code only when a condition is `True`.

```python
if condition:
    # this code only runs if condition is True
```

Notice two things: the colon `:` at the end of the `if` line, and the indentation (four spaces, typically) on the line below. Indentation is not optional in Python; it is how Python knows which code belongs inside the `if` block.


````
age = 20

if age >= 18:
    print("You are an adult.")

print("This line always runs, regardless of the condition.")
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Forgetting the colon | `if age >= 18` | `SyntaxError` | `if age >= 18:` |
| Forgetting to indent | `if age >= 18:` then unindented code | `IndentationError` | Indent the code block consistently |
| Using `=` instead of `==` | `if age = 18:` | `SyntaxError` | `if age == 18:` |
| Mixing tabs and spaces | Inconsistent indentation | `IndentationError` or silent bugs | Stick to spaces consistently (Jupyter does this automatically) |


## Step 2: Adding `else`

`else` provides a fallback: code that runs when the `if` condition is `False`.


````
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
````

## Step 3: Adding `elif`

`elif` (short for "else if") lets you check multiple conditions in sequence. Python checks each condition in order and runs the first one that is `True`, then skips the rest.


````
score = 72

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"

print("Grade:", grade)
````

## Step 5: Nested If Statements

You can place an `if` statement inside another `if` statement. This is called nesting, and it is useful when a decision depends on another decision.


````
stock_count = 5
is_popular = True

if stock_count > 0:
    if is_popular and stock_count < 10:
        print("Low stock warning: reorder soon.")
    else:
        print("In stock.")
else:
    print("Out of stock.")
````

## Fun Facts

- Python deliberately uses indentation instead of curly braces `{}` to define code blocks, a design choice meant to force consistently readable code across every Python project.
- The word "conditional" in programming comes directly from formal logic, the same branch of mathematics behind the `and`, `or`, and `not` operators you met in Notebook 3.
- Python also supports a compact one-line conditional called a ternary expression: `status = "Adult" if age >= 18 else "Minor"`, useful once you get comfortable with the full syntax.


````
# Fun fact demo — a ternary expression (one-line if/else)
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)
````

## Quick Recap

- `if condition:` runs a block of code only when the condition is `True`.
- `else:` provides a fallback for when the condition is `False`.
- `elif:` checks additional conditions in order, stopping at the first `True` match.
- Combine conditions with `and`, `or`, `not` from Notebook 3.
- Indentation is required in Python; it defines which code belongs to which block.
- You can nest `if` statements inside one another for layered decisions.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Basic if | `if condition:` | `if age >= 18:` |
| If / else | `if ... else:` | `if x > 0: ... else: ...` |
| Multiple conditions | `if ... elif ... else:` | `if x > 90: ... elif x > 60: ... else: ...` |
| Combine conditions | `and`, `or`, `not` | `if a and b:` |
| Nested if | `if:` inside `if:` | See Step 5 above |
| One-line conditional | `a if condition else b` | `"Yes" if x > 0 else "No"` |
