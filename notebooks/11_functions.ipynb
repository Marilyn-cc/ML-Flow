## Welcome Back

You have written a lot of code across the last ten notebooks, and some of it has probably felt repetitive. Functions solve exactly that problem. A function is a named, reusable block of code that you define once and can call as many times as you like, with different inputs each time.


## Why Does This Matter?

Imagine you need to calculate a 10% discount in five different places throughout a larger project. Copying the same formula five times means five places to fix if the discount logic ever changes. A function lets you write the logic once:

```python
def apply_discount(price):
    return price * 0.9
```

Now every part of your project calls `apply_discount(price)`, and if the discount rate ever changes, you update it in exactly one place. This is one of the biggest productivity shifts in this entire course.

## Step 2: Parameters — Passing Information In

Parameters let a function accept input, so it can behave differently each time you call it.


````
def greet(name):
    print(f"Hello, {name}!")

greet("Amina")
greet("Brian")
````

You can accept multiple parameters, separated by commas.


````
def describe_product(name, price):
    print(f"{name} costs KES {price}")

describe_product("Rice", 150)
describe_product("Milk", 65)
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Forgetting the colon after the parentheses | `def greet()` | `SyntaxError` | `def greet():` |
| Forgetting to call the function | Defining `greet()` but never writing `greet()` | Nothing happens, no error either | Remember to actually call the function |
| Forgetting `return` and expecting output | `def add(a, b): print(a + b)` used as `x = add(2, 3)` | `x` becomes `None` | Use `return a + b` if you want to store the result |
| Providing arguments in the wrong order | `describe_product(150, "Rice")` | Runs, but with mismatched values | Match the parameter order, or use named arguments |


## Step 3: Returning a Value

`print()` only displays a value; it does not let you reuse it elsewhere. `return` sends a value back out of the function so you can store it in a variable and use it later.


````
def apply_discount(price):
    return price * 0.9

original_price = 500
discounted_price = apply_discount(original_price)

print("Original:", original_price)
print("Discounted:", discounted_price)
````

## Step 4: Default Parameter Values

You can give a parameter a default value, used automatically if the caller does not provide one.


````
def apply_discount(price, discount_rate=0.10):
    return price * (1 - discount_rate)

print(apply_discount(1000))            # uses the default 10%
print(apply_discount(1000, 0.25))      # overrides with 25%
````

## Step 5: Multiple Return Values

A function can return more than one value at once. Behind the scenes, Python packages them into a tuple, exactly the structure you met in Notebook 6.


````
def get_price_range(prices):
    return min(prices), max(prices)

lowest, highest = get_price_range([150, 65, 300, 90])
print("Lowest:", lowest)
print("Highest:", highest)
````

## Step 6: Variable Scope

A variable created inside a function is called **local**; it only exists inside that function. A variable created outside any function is called **global**, and can be read from anywhere, though modifying it from inside a function requires special care that goes beyond this notebook.


````
tax_rate = 0.16   # a global variable

def calculate_tax(price):
    tax = price * tax_rate   # tax is local, tax_rate is read from the global scope
    return tax

print(calculate_tax(1000))

# Uncomment the line below to see the error: 'tax' does not exist outside the function
# print(tax)
````

## Fun Facts

- Functions you write yourself work exactly the same way as built-in functions like `len()`, `print()`, and `sum()`, which are themselves just functions that Python provides for you.
- The practice of avoiding repeated code by wrapping logic in a function is often summarized as "Don't Repeat Yourself," commonly abbreviated as DRY, and it is one of the most quoted principles in all of software engineering.
- Later in this course, you will call functions written by other people constantly, such as `pd.read_csv()` from the Pandas library, which works using the exact same `def` and `return` mechanics you just learned.


## Quick Recap

- Define a function with `def name(parameters):`.
- Parameters let a function accept different inputs each time it is called.
- `return` sends a value back out, so you can store or reuse the result.
- Default parameter values are used automatically when the caller omits an argument.
- A function can return multiple values at once, packaged as a tuple.
- Variables created inside a function are local and do not exist outside it.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Define a function | `def name():` | `def greet():` |
| Function with a parameter | `def name(param):` | `def greet(name):` |
| Default parameter value | `def name(param=value):` | `def greet(name="friend"):` |
| Return a value | `return value` | `return price * 0.9` |
| Call a function | `name(arguments)` | `greet("Amina")` |
| Multiple return values | `return a, b` | `return min(x), max(x)` |
| Local variable | Created inside a function | Only exists inside that function |
