## Welcome Back

Lists and tuples store items in order, accessed by position. Dictionaries take a different approach: every value gets its own descriptive label, called a key. Once you see how natural this feels, you will start reaching for dictionaries constantly, especially once you start working with real datasets and JSON data from APIs.


## Why Does This Matter?

Compare these two ways of representing a single product:

```python
product_list = ["Rice", 150, True]          # Which value is which?
product_dict = {"name": "Rice", "price": 150, "in_stock": True}   # Instantly clear
```

The dictionary version is self-explaining. This is exactly how a single row of data is represented in almost every API response, configuration file, and JSON dataset you will encounter as a data scientist.


## Step 1: Creating a Dictionary

A dictionary is created using curly braces `{}`, with each item written as `key: value`.


````
product = {"name": "Rice", "price": 150, "in_stock": True}

print(product)
print(type(product))
````

## Step 2: Accessing Values

Instead of using a numeric index, you access dictionary values using their key.


````
product = {"name": "Rice", "price": 150, "in_stock": True}

print(product["name"])
print(product["price"])

# .get() is a safer alternative that avoids errors on missing keys
print(product.get("category"))          # returns None instead of crashing
print(product.get("category", "N/A"))   # returns a default value instead
````

### A common beginner trap: `KeyError`

Accessing a key that does not exist with square brackets raises an error.


````
product = {"name": "Rice", "price": 150}

# Uncomment the line below to see the error:
# print(product["category"])

# Safer approach:
print(product.get("category", "Not specified"))
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Using square brackets for a missing key | `product["category"]` | `KeyError` | Use `.get()` with a default value |
| Forgetting the colon between key and value | `{"name" "Rice"}` | `SyntaxError` | `{"name": "Rice"}` |
| Using a mutable type (like a list) as a key | `{["a"]: 1}` | `TypeError` | Use a string, number, or tuple as the key instead |
| Confusing `.values()` with `.keys()` | Expecting keys, getting values | Wrong data used | Double-check which method you called |


## Step 3: Updating a Dictionary

Dictionaries are mutable, just like lists. You can add new keys or update existing ones using the same square bracket syntax.


````
product = {"name": "Rice", "price": 150}

product["price"] = 145          # update an existing key
product["category"] = "Grains" # add a new key

print(product)
````

### Removing keys

| Method | What it does | Example |
|---|---|---|
| `del dict[key]` | Deletes a key permanently | `del product["category"]` |
| `.pop(key)` | Removes a key and returns its value | `product.pop("category")` |
| `.clear()` | Empties the entire dictionary | `product.clear()` |


````
product = {"name": "Rice", "price": 145, "category": "Grains"}

removed_value = product.pop("category")
print("Removed:", removed_value)
print(product)

del product["price"]
print(product)
````

## Step 4: Keys, Values, and Items

| Method | What it returns | Example |
|---|---|---|
| `.keys()` | All keys in the dictionary | `product.keys()` |
| `.values()` | All values in the dictionary | `product.values()` |
| `.items()` | Each key-value pair, as tuples | `product.items()` |
| `in` | Checks if a key exists | `"price" in product` |


````
product = {"name": "Rice", "price": 150, "in_stock": True}

print(product.keys())
print(product.values())
print(product.items())
print("price" in product)
print("category" in product)
````

## Step 5: Looping Through a Dictionary

You can loop through a dictionary's keys, values, or both. Don't worry if `for` loops feel new; you will master them fully in Notebook 10. For now, focus on the pattern.


````
product = {"name": "Rice", "price": 150, "in_stock": True}

for key in product:
    print(key, "->", product[key])

print("---")

for key, value in product.items():
    print(f"{key}: {value}")
````

## Fun Facts

- Since Python 3.7, dictionaries remember the order items were added, which was not guaranteed in older versions of the language.
- Dictionary keys must be immutable, which is exactly why tuples (but not lists) can be used as keys.
- Almost every response from a real-world web API arrives as JSON, which converts directly into Python dictionaries and lists once loaded.


## Quick Recap

- Dictionaries store key-value pairs using curly braces `{}`.
- Access values with `dict[key]`, or more safely with `.get(key, default)`.
- Dictionaries are mutable: update or add keys directly, remove with `.pop()` or `del`.
- `.keys()`, `.values()`, and `.items()` let you inspect and loop through a dictionary.
- The `in` keyword checks whether a key exists.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Create a dictionary | `{key: value}` | `{"name": "Rice"}` |
| Access a value | `dict[key]` | `product["name"]` |
| Safe access | `.get(key, default)` | `product.get("size", "N/A")` |
| Update or add a key | `dict[key] = value` | `product["price"] = 150` |
| Remove a key | `del dict[key]` / `.pop(key)` | `product.pop("price")` |
| Get all keys | `.keys()` | `product.keys()` |
| Get all values | `.values()` | `product.values()` |
| Get key-value pairs | `.items()` | `product.items()` |
| Check a key exists | `key in dict` | `"price" in product` |
| Number of pairs | `len(dict)` | `len(product)` |
