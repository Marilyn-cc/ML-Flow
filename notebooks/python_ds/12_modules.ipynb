## Welcome Back

You now know how to write your own functions. Modules take that same idea one level up: entire files full of ready-made functions, written by other people, that you can bring into your own code with a single line. This is exactly how you will use NumPy, Pandas, Matplotlib, and every other library in the rest of this course.


## Why Does It Matter?

Imagine needing to calculate a square root, generate a random number, or work with dates. You could write all of that logic yourself, but why would you? Python ships with a huge standard library of modules covering exactly these needs, and thousands more modules are available from the wider Python community, including NumPy and Pandas, which you will meet very soon in this course.


## Step 1: Importing a Module

The `import` keyword brings a module's tools into your code.


````
import math

print(math.sqrt(16))
print(math.pi)
print(math.ceil(4.2))
print(math.floor(4.8))
````

Notice the pattern: `module_name.function_name()`. The dot connects the module to the specific tool you want to use inside it.

## Step 2: Using an Alias

Many modules have a commonly agreed-upon short alias, created using `as`. You will use this constantly once you start working with NumPy (`np`) and Pandas (`pd`).


````
import math as m

print(m.sqrt(25))
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Forgetting to import before using | `math.sqrt(4)` with no `import math` | `NameError` | Add `import math` at the top |
| Misspelling a module name | `import maths` | `ModuleNotFoundError` | The correct name is `math` |
| Forgetting the module prefix | `sqrt(4)` instead of `math.sqrt(4)` | `NameError` | Use `module_name.function()`, unless you imported the function directly |
| Re-importing unnecessarily | Repeating `import math` many times | Not an error, just unnecessary | One `import` per module, usually at the top of the file |


## Step 3: Importing Specific Functions

If you only need one or two tools from a module, you can import them directly, avoiding the need for the module prefix each time.


````
from math import sqrt, pi

print(sqrt(49))
print(pi)
````

## Step 4: The `random` Module

Another very common built-in module is `random`, useful for generating random numbers, which you will use again once you reach the NumPy notebooks later in this course.


````
import random

print(random.randint(1, 10))       # a random whole number between 1 and 10
print(random.choice(["Milk", "Bread", "Rice"]))   # a random item from a list
print(random.random())              # a random decimal between 0 and 1
````

## Step 5: The `datetime` Module

The `datetime` module handles dates and times, another skill you will use constantly once working with real datasets.


````
from datetime import datetime

now = datetime.now()
print(now)
print(now.year)
print(now.month)
print(now.day)
````

## Fun Facts

- Python's standard library, the collection of built-in modules that ships with every Python installation, is often described as "batteries included," meaning most common tasks already have a ready-made tool.
- NumPy, Pandas, Matplotlib, and every other tool used in the rest of this course are simply modules, installed separately and imported using the exact same `import` pattern you just practiced.
- The alias conventions `np` for NumPy and `pd` for Pandas are not required by Python, but they are followed so consistently across the entire data science community that using anything else would confuse most other developers reading your code.


## Quick Recap

- A module is a file full of reusable functions and tools, brought into your code with `import`.
- Use `module.function()` to call a tool from an imported module.
- `import module as alias` creates a shorter name for convenience.
- `from module import function` imports a specific tool directly, without needing the module prefix.
- Common built-in modules include `math`, `random`, and `datetime`.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Import a module | `import module` | `import math` |
| Import with an alias | `import module as alias` | `import numpy as np` |
| Import a specific function | `from module import function` | `from math import sqrt` |
| Use a module's function | `module.function()` | `math.sqrt(16)` |
| Random integer | `random.randint(a, b)` | `random.randint(1, 10)` |
| Random choice from a list | `random.choice(list)` | `random.choice(fruits)` |
| Current date and time | `datetime.now()` | `datetime.now()` |
