## Step 2: Indexing and Slicing Lists

Lists use the exact same indexing and slicing rules you already learned for strings in Notebook 4.

```
 "Milk"  "Bread"  "Rice"  "Sugar"  "Tea Leaves"
   0        1        2       3          4
  -5       -4       -3      -2         -1
```


````
products = ["Milk", "Bread", "Rice", "Sugar", "Tea Leaves"]

print(products[0])      # first item
print(products[-1])     # last item
print(products[1:3])    # a slice: items at index 1 and 2
print(products[:2])     # first two items
print(products[-2:])    # last two items
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Assuming indexing starts at 1 | `products[1]` to get the first item | Actually returns the **second** item | Remember indexing starts at `0` |
| Going out of range | `products[10]` on a 5-item list | `IndexError` | Check `len(products)` first |
| Forgetting slicing excludes the stop index | `products[0:2]` expecting 3 items | Only returns items at index 0 and 1 | Remember `stop` is not included |
| Confusing `.append()` with `.insert()` | Using `.append()` to add at a specific position | Always adds to the **end** | Use `.insert(index, value)` for a specific position |


## Step 3: Lists Are Mutable

Unlike strings, lists **can** be changed after creation. This is called being "mutable."


````
products = ["Milk", "Bread", "Rice"]

# Update an item directly by index
products[1] = "Brown Bread"
print(products)
````

### Adding items

| Method | What it does | Example |
|---|---|---|
| `.append(x)` | Adds `x` to the end | `products.append("Sugar")` |
| `.insert(i, x)` | Inserts `x` at index `i` | `products.insert(0, "Tea")` |
| `.extend(list)` | Adds all items from another list | `products.extend(["Eggs", "Butter"])` |


````
products = ["Milk", "Bread", "Rice"]

products.append("Sugar")
print(products)

products.insert(0, "Tea Leaves")
print(products)

products.extend(["Eggs", "Butter"])
print(products)
````

### Removing items

| Method | What it does | Example |
|---|---|---|
| `.remove(x)` | Removes the first matching item `x` | `products.remove("Bread")` |
| `.pop()` | Removes and returns the last item (or a given index) | `products.pop()` |
| `del list[i]` | Deletes the item at a specific index | `del products[0]` |
| `.clear()` | Empties the entire list | `products.clear()` |


````
products = ["Tea Leaves", "Milk", "Bread", "Rice", "Sugar"]

products.remove("Bread")
print(products)

last_item = products.pop()
print("Removed:", last_item)
print(products)

del products[0]
print(products)
````

## Step 4: Useful List Methods

| Method | What it does | Example |
|---|---|---|
| `.sort()` | Sorts the list in place, ascending | `numbers.sort()` |
| `.sort(reverse=True)` | Sorts in descending order | `numbers.sort(reverse=True)` |
| `sorted(list)` | Returns a new sorted list, original unchanged | `sorted(numbers)` |
| `.reverse()` | Reverses the order in place | `numbers.reverse()` |
| `.count(x)` | Counts how many times `x` appears | `numbers.count(10)` |
| `.index(x)` | Finds the index of the first match | `numbers.index(10)` |
| `in` | Checks if a value exists in the list | `10 in numbers` |
| `sum()`, `min()`, `max()` | Work directly on numeric lists | `sum(numbers)` |


````
prices = [150, 65, 300, 90, 65, 220]

prices.sort()
print("Sorted:", prices)

prices.sort(reverse=True)
print("Descending:", prices)

print("Count of 65:", prices.count(65))
print("Sum:", sum(prices))
print("Minimum:", min(prices))
print("Maximum:", max(prices))
print("Is 300 in the list?", 300 in prices)
````

## Fun Facts

- Python lists can grow or shrink freely; you never need to declare a fixed size in advance, unlike arrays in many other languages.
- A list can contain other lists inside it, called a nested list, which is exactly how a simple table or grid of data can be represented before you meet Pandas DataFrames.
- The `in` keyword you used to check membership works on strings too: `"a" in "cat"` returns `True`.


````
# Fun fact demo — a nested list, like a tiny table
table = [
    ["Milk", 65.5],
    ["Bread", 55],
    ["Rice", 150]
]

print(table[0])       # first row
print(table[0][0])    # first row, first column
print(table[2][1])    # third row, second column
````

## Quick Recap

- Lists are created with square brackets and can hold multiple values of any type.
- Indexing and slicing work the same way as strings, starting at `0`.
- Lists are mutable: you can update, add, and remove items after creation.
- Common methods: `.append()`, `.insert()`, `.remove()`, `.pop()`, `.sort()`.
- Built-in functions `sum()`, `min()`, `max()`, and `len()` work directly on lists.
- The `in` keyword checks whether a value exists inside a list.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Create a list | `[item1, item2]` | `["Milk", "Bread"]` |
| Access an item | `list[index]` | `fruits[0]` |
| Slice a list | `list[start:stop]` | `fruits[1:3]` |
| Update an item | `list[index] = value` | `fruits[0] = "Kiwi"` |
| Add to the end | `.append(x)` | `fruits.append("Kiwi")` |
| Insert at a position | `.insert(i, x)` | `fruits.insert(0, "Kiwi")` |
| Add multiple items | `.extend(list)` | `fruits.extend(["Kiwi","Fig"])` |
| Remove by value | `.remove(x)` | `fruits.remove("Kiwi")` |
| Remove by position | `.pop()` / `del list[i]` | `fruits.pop()` |
| Sort ascending | `.sort()` | `fruits.sort()` |
| Sort descending | `.sort(reverse=True)` | `fruits.sort(reverse=True)` |
| Check membership | `x in list` | `"Kiwi" in fruits` |
| List length | `len(list)` | `len(fruits)` |
