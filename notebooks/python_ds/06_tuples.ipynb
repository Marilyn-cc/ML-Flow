## Welcome Back

This notebook is a short one, and that is by design. Tuples are essentially lists with one important twist: once created, they cannot be changed. That single difference turns out to matter quite a lot, and you are about to see why.


## Why Does This Matter?

Some real-world data should never change once it is recorded. Consider a single GPS coordinate:

```python
nairobi_location = (-1.286389, 36.817223)
```

Latitude and longitude are a fixed pair. It would not make sense for code elsewhere to accidentally overwrite one half of that pair. A tuple protects data like this from accidental changes, which is exactly the kind of safety net that matters once your code grows larger and is used by other people.


## Step 1: Creating a Tuple

A tuple is created using parentheses `()` instead of square brackets.


````
coordinates = (-1.286389, 36.817223)
print(coordinates)
print(type(coordinates))
````

Tuples support indexing and slicing exactly like lists.


````
dimensions = (1920, 1080, "landscape")

print(dimensions[0])     # width
print(dimensions[1])     # height
print(dimensions[2])     # orientation
print(dimensions[0:2])   # slicing works too
print(len(dimensions))
````

## Step 2: Tuples Are Immutable

Unlike lists, once a tuple is created, you cannot change, add, or remove its items.


````
coordinates = (-1.286389, 36.817223)

# Uncomment the line below to see the error:
# coordinates[0] = 0

print(coordinates)   # unchanged, and that's the whole point
````

This might feel limiting at first, but it is actually a feature. Immutability means:

- The data cannot be accidentally modified elsewhere in your code
- Tuples are slightly faster than lists for read-only data
- Tuples can be used as dictionary keys later on, which lists cannot


## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Trying to modify a tuple | `point[0] = 5` | `TypeError` | Create a new tuple instead |
| Forgetting the comma in a single-item tuple | `single = (5)` | This is just an `int`, not a tuple! | `single = (5,)` with a trailing comma |
| Assuming tuples can use `.append()` | `point.append(3)` | `AttributeError` | Tuples do not support list-only methods |
| Confusing tuple parentheses with function calls | `(5)` looks like grouping, not a tuple | Easy to mix up | Always include the trailing comma for one-item tuples |


````
not_a_tuple = (5)
print(type(not_a_tuple))   # int, surprising but true!

actual_tuple = (5,)
print(type(actual_tuple))  # tuple, thanks to the trailing comma
````

## Step 3: Unpacking Tuples

One of the most useful tuple features is unpacking: assigning each item in a tuple to its own variable in a single line.


````
location = ("Nairobi", -1.286389, 36.817223)

city, latitude, longitude = location

print(city)
print(latitude)
print(longitude)
````

This is the same multiple-assignment trick you briefly saw in Notebook 1, and it works because Python automatically treats `a, b, c = 1, 2, 3` as tuple unpacking behind the scenes.

## Step 4: When to Use a Tuple vs a List

| Situation | Use a... | Why |
|---|---|---|
| A shopping list that will change | List | You need `.append()`, `.remove()`, etc. |
| A fixed coordinate pair | Tuple | Latitude and longitude should not change |
| Rows of data you are only reading | Tuple | Protects the data from accidental edits |
| A collection you will sort or filter | List | Lists support in-place sorting and modification |
| Function results returning multiple values | Tuple | This is Python's standard convention |

A simple rule of thumb: if the data will grow, shrink, or be edited, use a list. If it represents a fixed, complete record, a tuple is often the better and safer choice.


````
# A function that returns multiple values actually returns a tuple
def get_min_max(numbers):
    return min(numbers), max(numbers)

result = get_min_max([15, 3, 42, 8])
print(result)
print(type(result))

# You can unpack the result directly
lowest, highest = get_min_max([15, 3, 42, 8])
print(lowest, highest)

# Don't worry about writing your own functions yet, you'll master that in Notebook 11!
````

## Fun Facts

- Tuples are commonly used to return multiple values from a function, which you will use constantly once you start writing your own functions.
- Because tuples are immutable, Python can use them as dictionary keys, something lists are never allowed to do, which you will see in the next notebooks.
- Tuple unpacking also works with the star operator for grabbing "the rest": `first, *rest = (1, 2, 3, 4)` gives `first = 1` and `rest = [2, 3, 4]`.


````
# Fun fact demo — star unpacking
first, *rest = (1, 2, 3, 4)
print(first)
print(rest)
````

## Quick Recap

- Tuples are created with parentheses and behave like lists, except they are immutable.
- Indexing and slicing work the same as lists and strings.
- A single-item tuple needs a trailing comma: `(5,)`.
- Unpacking lets you assign each tuple item to its own variable in one line.
- Use tuples for fixed, protected data; use lists when the collection needs to change.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Create a tuple | `(item1, item2)` | `("Rice", 150)` |
| Single-item tuple | `(item,)` | `(5,)` |
| Access an item | `tuple[index]` | `point[0]` |
| Slice a tuple | `tuple[start:stop]` | `point[0:2]` |
| Unpack a tuple | `a, b = my_tuple` | `x, y = point` |
| Star unpacking | `first, *rest = my_tuple` | `a, *b = (1,2,3)` |
| Tuple length | `len(tuple)` | `len(point)` |
| Attempted modification | Not allowed | Raises `TypeError` |
