## Step 1: Writing to a File

Let's start by creating a file, since we need something to read afterward. The `with open(...)` pattern is the standard, safe way to work with files in Python: it automatically closes the file for you, even if something goes wrong.

```python
with open("filename.txt", "w") as file:
    file.write("some text")
```

The second argument, `"w"`, means "write mode." Be careful: write mode completely replaces the file's existing content, if any.


````
with open("shop_notes.txt", "w") as file:
    file.write("Shop opened at 8:00 AM.\n")
    file.write("Delivery received: 50 units of rice.\n")

print("File written successfully.")
````

Notice `\n`, the newline character. Without it, everything you write would end up on a single line.

## Step 2: Reading a File

Reading uses the same `with open(...)` pattern, but with `"r"` for "read mode."


````
with open("shop_notes.txt", "r") as file:
    content = file.read()

print(content)
````

### Reading Line by Line

Often it is more useful to process a file one line at a time, which is exactly what you will do constantly once loading real datasets row by row.


````
with open("shop_notes.txt", "r") as file:
    for line in file:
        print("Line:", line.strip())   # .strip() removes the trailing newline character
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Using `"w"` mode when you meant to keep existing content | `open("log.txt", "w")` on an existing file | Existing content is erased | Use `"a"` (append mode) instead |
| Forgetting to strip newline characters | Printing raw lines from a file | Extra blank lines or trailing `\n` in output | Use `.strip()` on each line |
| Not using `with` | `file = open(...)` without closing it | The file may stay locked or unsaved | Always use `with open(...) as file:` |
| Trying to read a file that does not exist | `open("missing.txt", "r")` | `FileNotFoundError` | Check the file exists first, or use a `try`/`except` block |


## Step 3: Appending to a File

Append mode, `"a"`, adds new content to the end of a file without erasing what is already there.


````
with open("shop_notes.txt", "a") as file:
    file.write("Closing stock count completed.\n")

with open("shop_notes.txt", "r") as file:
    print(file.read())
````

## Step 4: Handling a Missing File Safely

Trying to read a file that does not exist raises a `FileNotFoundError`. You can handle this gracefully using `try` and `except`, so your program does not crash unexpectedly.


````
try:
    with open("does_not_exist.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("That file does not exist yet.")
````

## Fun Facts

- The `with` statement you have been using throughout this notebook is called a "context manager," and it guarantees the file gets closed properly, even if an error occurs partway through.
- Every CSV file you will load in the upcoming Pandas notebooks is, underneath everything, just a plain text file, exactly like the ones you just created, but formatted with commas separating each value.
- Beyond plain text files, Python can also read and write more structured formats like JSON and CSV directly, using additional built-in modules you will meet as this course continues.


## Quick Recap

- `with open(filename, mode) as file:` safely opens a file and closes it automatically.
- `"w"` mode writes and erases existing content; `"a"` mode appends without erasing.
- `"r"` mode reads; use `.read()` for the whole file, or loop through the file object for line-by-line access.
- `.strip()` removes trailing newline characters when reading lines.
- Wrap risky file operations in `try`/`except` to handle missing files gracefully.
