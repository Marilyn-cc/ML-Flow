## Step 1: Creating and Combining Strings

You already know the basics from Notebook 2. Let's go a bit deeper.


````
greeting = "Hello"
name = "Amina"

# Concatenation (joining strings with +)
message = greeting + ", " + name + "!"
print(message)

# Repetition (repeating a string with *)
line = "-" * 20
print(line)
````

### f-strings: the cleaner way to combine text

You met f-strings briefly in earlier notebooks. They let you insert variables directly inside a string using curly braces `{}`.


````
name = "Amina"
age = 24

# Old way
message_1 = "My name is " + name + " and I am " + str(age) + " years old."

# f-string way (cleaner, and handles type conversion automatically)
message_2 = f"My name is {name} and I am {age} years old."

print(message_1)
print(message_2)
````

Notice that in the f-string version, you did not need to convert `age` to a string with `str()`. f-strings handle that automatically. This is why most Python developers prefer f-strings for formatting text.

## Step 2: String Indexing

Every character in a string has a position, called an **index**, starting from `0`, not `1`.

```
 P   y   t   h   o   n
 0   1   2   3   4   5
-6  -5  -4  -3  -2  -1
```

Notice there are also negative indices, which count backward from the end of the string.


````
word = "Python"

print(word[0])    # first character
print(word[5])    # last character
print(word[-1])   # also the last character
print(word[-6])   # also the first character
````

### A common beginner trap: `IndexError`

If you try to access an index that does not exist, Python raises an error.


````
word = "Python"

# Uncomment the line below to see the error:
# print(word[10])

print(len(word))   # use len() to check how many characters a string has
````

## Common Beginner Mistakes

| Mistake | Example | What happens | Fix |
|---|---|---|---|
| Assuming indexing starts at 1 | `word[1]` to get the first letter | Actually returns the **second** letter | Remember indexing starts at `0` |
| Going out of range | `word[100]` on a short string | `IndexError` | Check `len(word)` first |
| Trying to modify a character directly | `word[0] = "J"` | `TypeError`, strings are immutable | Build a new string instead |
| Forgetting string methods return a new string | `word.upper()` without reassigning | Original `word` stays unchanged | `word = word.upper()` |


## Step 3: String Slicing

Slicing lets you extract a **portion** of a string using `string[start:stop]`. The `start` index is included, the `stop` index is **not**.


````
word = "Python"

print(word[0:3])    # characters at index 0, 1, 2
print(word[2:])     # from index 2 to the end
print(word[:4])     # from the beginning up to (not including) index 4
print(word[:])      # the whole string
print(word[-3:])    # the last three characters
````

## Step 4: Useful String Methods

Strings come with many built-in methods for cleaning and transforming text. Here are the ones you will use most often.

| Method | What it does | Example | Result |
|---|---|---|---|
| `.upper()` | Converts to uppercase | `"data".upper()` | `"DATA"` |
| `.lower()` | Converts to lowercase | `"DATA".lower()` | `"data"` |
| `.title()` | Capitalizes each word | `"amina otieno".title()` | `"Amina Otieno"` |
| `.strip()` | Removes leading/trailing whitespace | `"  data  ".strip()` | `"data"` |
| `.replace(a, b)` | Replaces one substring with another | `"data".replace("a", "o")` | `"doto"` |
| `.split(sep)` | Splits a string into a list | `"a,b,c".split(",")` | `["a", "b", "c"]` |
| `.join(list)` | Joins a list into a string | `"-".join(["a","b"])` | `"a-b"` |
| `.startswith(x)` | Checks if string starts with `x` | `"data".startswith("da")` | `True` |
| `.find(x)` | Finds the index of a substring | `"data".find("t")` | `2` |


````
messy_name = "  amina OTIENO  "

print(messy_name.strip())
print(messy_name.strip().title())
print(messy_name.upper())
print(messy_name.lower())
````

````
sentence = "I love Data Science"

words = sentence.split(" ")
print(words)

rejoined = "-".join(words)
print(rejoined)

print(sentence.replace("love", "enjoy"))
print(sentence.startswith("I love"))
print(sentence.find("Data"))
````

## Fun Facts

- Strings in Python are **immutable**, meaning once created, they cannot be changed in place. Every method like `.upper()` or `.strip()` actually returns a brand new string.
- The trick `string[::-1]` reverses a string using slicing, and it is one of the most shared one-liners among Python beginners.
- Python strings fully support emoji and non-English characters, including Swahili text, thanks to built-in Unicode support.


## Quick Recap

- Strings can be combined with `+`, repeated with `*`, and formatted cleanly with f-strings.
- Indexing accesses a single character; indexing starts at `0`.
- Slicing extracts a range of characters: `string[start:stop]`.
- Common cleaning methods: `.strip()`, `.upper()`, `.lower()`, `.title()`, `.replace()`.
- `.split()` breaks a string into a list; `.join()` does the reverse.
- Strings are immutable: methods always return a new string, so you must reassign the result.

## Cheat Sheet

| Concept | Syntax | Example |
|---|---|---|
| Combine strings | `a + b` | `"Hi " + "there"` |
| Repeat a string | `a * n` | `"-" * 10` |
| Format with variables | `f"{variable}"` | `f"Age: {age}"` |
| Access a character | `string[index]` | `word[0]` |
| Slice a range | `string[start:stop]` | `word[0:3]` |
| Reverse a string | `string[::-1]` | `"abc"[::-1]` |
| Remove whitespace | `.strip()` | `"  hi  ".strip()` |
| Change case | `.upper() / .lower() / .title()` | `"hi".upper()` |
| Replace text | `.replace(old, new)` | `"hi".replace("h","H")` |
| Split into a list | `.split(sep)` | `"a,b".split(",")` |
| Join a list into a string | `sep.join(list)` | `"-".join(["a","b"])` |
| String length | `len(string)` | `len("hello")` |
